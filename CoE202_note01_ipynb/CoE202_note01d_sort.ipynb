## 1. Implementing sorting function
Let's use 'def' to define a sorting function.

```
# this is just an annotation
import numpy as np # this is for importing numpy library (and we will use abbreviation np for that)
```

```
# define an array-sorting function
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)
```

```
# use the sorting function
print(quicksort([3,6,8,10,1,2,1]))
```
